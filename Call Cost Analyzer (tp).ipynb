```python
import locale
locale.setlocale(locale.LC_ALL, "")

# Function for output
def BuildOutput(categoryList, lengthCall, costCall):
    print('{:<20s} {:>10s} {:>10s}'.format("Category", "Length", "Cost"))

# Function to calculate total call cost
def CalculateCallCost(costCall):
    return costCall

# Function to check if a string is a floating point number
def IsFloat(index):
    try:
        float(index)
        return True
    except:
        return False

categoryList = ["Daytime", "Evening", "Off-peak"]
ratePerMins = [0.07, 0.12, 0.05]
lengthCall = [0, 0, 0]
costCall = [0, 0, 0]

index = 0

while index != 4:
    try:
        # Input option for time call
        index = int(input("Select the time of the phone call\n\
1. 6am to 5:59pm\n\
2. 6pm to 11:59pm\n\
3. 12am to 5:59am\n\
4. End\n"))

        # While loop to reject values not in the range (1-4)
        while index != 1 and index != 2 and index != 3 and index != 4:
            print("Must enter 1, 2, 3, or 4")
            index = int(input("Select the time of the phone call\n\
1. 6am to 5:59pm\n\
2. 6pm to 11:59pm\n\
3. 12am to 5:59am\n\
4. End\n"))

        if index == 4:
            print("Program has ended")  # If index input = 4, exit the loop and end the program
            break

        # While loop for index (1-3)
        while index == 1 or index == 2 or index == 3:
            # Insert minutes length
            minutes = float(input("Enter the length of the call in minutes: "))
            
            if minutes > 0:
                index = index - 1
                lengthCall[index] += minutes
                costCall[index] = lengthCall[index] * ratePerMins[index]
                totalCost = CalculateCallCost(costCall)

            if minutes < 0:  # Statement to make sure no negative input
                print("Call length must be a positive number")
                continue

            index = int(input("Select the time of the phone call\n\
1. 6am to 5:59pm\n\
2. 6pm to 11:59pm\n\
3. 12am to 5:59am\n\
4. End\n"))

        # Loop to avoid the error of inserting invalid options
        while index != 1 and index != 2 and index != 3 and index != 4:
            print("Must enter 1, 2, 3, or 4")
            index = int(input("Select the time of the phone call\n\
1. 6am to 5:59pm\n\
2. 6pm to 11:59pm\n\
3. 12am to 5:59am\n\
4. End\n"))

        BuildOutput(categoryList, lengthCall, costCall)  # Define output

        for x in range(3):  # Show the 3 options of the category list
            if lengthCall[x] != 0:
                # Print out the result
                print('{:<20s} {:>10.1f} {:>10} '.format(categoryList[x], lengthCall[x], locale.currency(totalCost[x])))

        print('{:<20s} {:>10.1f} {:>10} '.format("Total", sum(lengthCall), locale.currency(sum(totalCost))))

    except ValueError:
        print("Input must be a decimal or an integer")
        retry = input("Press 0 to re-enter or any other key to exit: ")
        if retry != "0":
            print("Program has ended") 
            print("Please refresh program again!")
            break
```
